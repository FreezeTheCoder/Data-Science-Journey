## Starting the Data Cleaning Assignment with the Instructions provided

### 1) Converting the Height and Weight into numerical forms.

* Some of the height values are in cm and some are in feet/inches. (Example: {'170cm', '6'2"', etc.,}
<br></br>-> We will be defining a function to split the feet and inches and apply a lambda function to iterate through the data values and convert them based on what they end with.
<br></br>
* Some of the weight values are in kgs and some are in lbs. (Example: {'94kgs','85lbs', etc.,}
<br></br>-> We will be using the same procedure as what we did for height.

```python
def extract_height_ft_in(height_str):
    try:
        feet, inches = height_str.split("'")
        return int(feet) * 30.48 + int(inches) * 2.54
    except ValueError:
        return None

df['Height'] = df['Height'].apply(lambda x: float(x.replace('cm', '')) if x.endswith('cm') else extract_height_ft_in(x))

df['Height'].head()
```

```python
def extract_weight_lbs(weight_str):
    try:
        weight_lbs = float(weight_str.replace('lbs', ''))
        return weight_lbs*0.453592
    except ValueError:
        return None

df['Weight'] = df['Weight'].apply(lambda x: float(x.replace('kg', '')) if x.endswith('kg') else extract_weight_lbs(x))

df['Weight'].head()
```

```python
df.head()
```

### 2) Removing the escape sequences from columns.

* We will use string replace function and regex for this operation.

```python
df['Club'] = df['Club'].str.replace(r'\s*', '', regex=True)

df['Club'].head()
```

### 3) Converting the {'Value', 'Wage', 'Release Clause'} columns into numericals

* These columns have string values starting with '€' symbol. We will be replacing it.
* These columns have string values ending 'M' or 'K'. We will be converting them into numerics using a lambda and endswith function.

```python
money_cols = ['Value', 'Wage', 'Release Clause']

def clean_money(money_str):
    units = money_str[-1]
    value = money_str[:-1] if units in ('M','K') else money_str
    return float(value)*1000000 if units == 'M' else (float(value)*1000 if units == 'K' else None)

for col in money_cols:
    df[col]=df[col].str.strip('€')
    df[col]=df[col].apply(clean_money)
```

```python
df[money_cols].head()
```

## 4) Removing the ★ from columns that have it

* Some columns have star symbol. We will remove it using replace function and regex.

```python
star_cols = ['W/F', 'SM', 'IR']

for col in star_cols:
    df[col]=df[col].str.replace('★', '', regex=True)
```

```python
df[star_cols].head()
```

### 5) Converting the Hits column to numeric

* Some hits have 'K' in them, meaning they are thousands. We will be converting them to numeric using lambda function.

```python
def clean_hits(hits_str):
    try:
        if isinstance(hits_str, str):
            hits_str = hits_str.strip()
            if hits_str.endswith('K'):
                value = float(hits_str[:-1]) * 1000
            else:
                value = float(hits_str)
            return int(value)
    except ValueError:
        return None

df['Hits'] = df['Hits'].apply(clean_hits)
```
